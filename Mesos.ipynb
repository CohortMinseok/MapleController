```python
import numpy as np
import pandas as pd
```

```python
class MesosStats:
    def __init__(self):
        self.mesos = 0
        self.incomes = []
        
    def add(self, price, amount=None, name=None):
        self.incomes.append({
            "name": name,
            "price": price,
            "amount": amount,
            "total": price * amount
        })
        
    def table(self, current_mesos=0):
        rows = []
        
        for income in self.incomes:
            rows.append({
                'Name': income['name'],
                "Price": income['price'].__format__(','),
                "Amount": income['amount'].__format__(','),
                'Total': income['total'].__format__(',')
            })
            
        rows.append({
                'Name': 'Total',
                'Total': self.total.__format__(',')
            })
        if current_mesos != 0:
            rows.append({
                    'Name': 'Current + Total',
                    'Total': (self.total + current_mesos).__format__(',')
                })
            
        return pd.DataFrame(rows)

    @property
    def total(self):
        return sum(map(lambda x: x['total'], self.incomes))
    
    def __str__(self):
        return self.total.__format__(",")
    def __repr__(self):
        return self.__str__()
```

```python
# Selling
mesos = MesosStats()

# Spell Trace
mesos.add(5000, 9000 + 9000 + 7676, name='Spell Trace')

# Cubic Blade
mesos.add(489998, 27, name='Cubic Blade')

# Cubic Chaos Blade
mesos.add(11900000, 8, name='Cubic Chaos Blade')

# Unrelenting Flame
mesos.add(1299999, 32, name='Unrelenting Flame')

# Forever Unrelenting Flame
mesos.add(10999996, 10, name='Forever Unrelenting Flame')

# Clean Slate Scroll 5%
mesos.add(5999999, 1, name='Clean Slate Scroll 5%')

# Clean Slate Scroll 10%
mesos.add(16999999, 3, name='Clean Slate Scroll 10%')

# Potential Scroll
mesos.add(289999, 7, name='Potential Scroll')

# Chaos Scroll 60%
mesos.add(399999, 5, name='Chaos Scroll 60%')

# Eternal Rebirth Flame
mesos.add(64999999, 1, name='Eternal Rebirth Flame')

mesos.table(530873100)
```
